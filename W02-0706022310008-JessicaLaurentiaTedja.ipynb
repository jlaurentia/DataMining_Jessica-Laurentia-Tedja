```python
# Calculate null values and percentage
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

# Create a DataFrame for columns with null values
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]


# Format the percentage with two decimal places and % symbol
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)


null_df
```

By calculating both **the count and percentage** of null values for each column, I can clearly identify which features have incomplete information and understand how significant the missing data is.

```python
#Handle Missing Values
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].median())
df['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].median())
df['land_size_m2'] = df['land_size_m2'].fillna(df['land_size_m2'].median())
df['building_size_m2'] = df['building_size_m2'].fillna(df['building_size_m2'].median())
df['floors'] = df['floors'].fillna(df['floors'].median())
```

Handling missing values in the numerical columns is done by filling them with the **median** of each respective column. The median is chosen because, unlike the mode, it better represents the central tendency of continuous numerical data and is less likely to be distorted when the values are spread out or skewed. Using the median ensures that the replacement values remain **realistic and balanced**, making the dataset more complete and consistent while preserving its overall distribution for further analysis or modeling.

```python
cat_cols = ['building_orientation']

for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown')
```

Missing values in the categorical column **building_orientation** are handled by replacing them with the label **"Unknown"**. This approach is used because categorical data cannot be filled with statistical measures such as the median or mean.

By assigning "Unknown" to the missing entries, the information that the value is not available is preserved without distorting the existing categories.

```python
# Mapping Facilities

def categorize_facilities_num(fac_str):
    if pd.isna(fac_str):   # if NaN
        return 0
    facilities = [f.strip().lower() for f in fac_str.split(',')]
    count = len(facilities)

    if 1 <= count <= 3:
        return 1
    elif 4 <= count <= 6:
        return 2
    elif count >= 7:
        return 3
    else:
        return 0

# Overwrite the original 'facilities' column with numeric categories
df['facilities'] = df['facilities'].apply(categorize_facilities_num)

# Display the updated DataFrame
df.head()
```

```python
# Mapping Certificate

def simplify_certificate_num(val):
    if pd.isna(val):
        return 0   # Lainnya
    val = val.lower()
    if "shm" in val or "sertifikat hak milik" in val:
        return 1   # SHM
    elif "hgb" in val or "hak guna bangunan" in val:
        return 2   # HGB
    elif "hp" in val or "hak pakai" in val:
        return 3   # HP
    else:
        return 0   # Lainnya

df['certificate'] = df['certificate'].apply(simplify_certificate_num)
df.head()
```

```python
cert_counts = df['certificate'].value_counts().sort_index()
cert_labels = {
    0: "Lainnya",
    1: "SHM",
    2: "HGB",
    3: "HP"
}
plt.figure(figsize=(8,6))
ax = cert_counts.plot(kind='bar', color='steelblue')

ax.set_xticks(range(len(cert_counts)))
ax.set_xticklabels([cert_labels[i] for i in cert_counts.index], rotation=0)

plt.title("Distribution of Certificate Types", fontsize=14)
plt.xlabel("Certificate Type", fontsize=12)
plt.ylabel("Number of Properties", fontsize=12)

for i, v in enumerate(cert_counts):
    ax.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

The bar plot shows the distribution of property certificate types in the dataset. Each bar represents one category: **Lainnya, SHM, HGB, dan HP**. The height of the bar indicates how many properties fall into that category, while the labels above each bar display the exact counts. This makes it easy to see which certificate type is the most common and how the others compare.

```python
#Handle Missing Values
df['year_built'] = df['year_built'].fillna(df['year_built'].median())
df.head()
```

```python
# Fill NaN in building_age with (2025 - year_built)

df['building_age'] = df['building_age'].fillna(2025 - df['year_built'])
df.head()
```

```python
# Display unique values for each column
for col in df.columns:
  print(f"Unique values for column '{col}':")
  print(df[col].unique())
  print("-" * 20)
```

```python
bins = [0, 5, 10, 15, 20, 25, df['building_age'].max()]
labels = ['0-5 years', '6-10 years', '11-15 years',
          '16-20 years', '21-25 years', '>25 years']
df['building_age'] = pd.cut(df['building_age'],
                            bins=bins,
                            labels=labels,
                            include_lowest=True)
df.head()
```

Binning is applied to the **building_age** column to make the data easier to interpret and analyze. Since building age is a continuous numerical variable, the raw values can be scattered and difficult to compare directly. By grouping the ages into ranges such as **0–5 years, 6–10 years, 11-15 years, 16-20 years, 21-25 years, or >25 years**, the information becomes more structured and meaningful.

This approach highlights patterns more clearly, for example showing whether newer or older houses dominate the dataset, and simplifies further analysis or visualization.

```python
# Building Age Distribution (Count Visualization)
age_counts = df['building_age'].value_counts().sort_index()
plt.figure(figsize=(10,6))
ax = age_counts.plot(kind='bar', color='steelblue')

plt.title("Distribution of Building Age", fontsize=14)
plt.xlabel("Building Age Category", fontsize=12)
plt.ylabel("Number of Properties", fontsize=12)
plt.xticks(rotation=0)
for i, v in enumerate(age_counts):
    ax.text(i, v + 0.5, str(v), ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
df['furnishing'] = df['furnishing'].replace('baru', 'Unknown')
df.loc[df['furnishing'].isna() & df['property_condition'].isin(['semi furnished', 'unfurnished']), 'furnishing'] = df['property_condition']
df.loc[df['furnishing'].isna() & ~df['property_condition'].isin(['semi furnished', 'unfurnished']), 'furnishing'] = 'Unknown'

df.loc[df['property_condition'].isna() & (df['furnishing'] == 'baru'), 'property_condition'] = df['furnishing']
df.loc[df['property_condition'].isna() & (df['furnishing'] != 'baru'), 'property_condition'] = 'Unknown'
df['property_condition'] = df['property_condition'].replace(['semi furnished', 'unfurnished'], 'Unknown')

print (df['property_condition'].value_counts())
print (df['furnishing'].value_counts())
```

```python
# Create New Feature

# Define encoding dictionaries
prop_enc = {
    "baru": 3,
    "bagus": 1,
    "bagus sekali": 2,
    "sudah renovasi": 2,
    "butuh renovasi": 0,
    "unknown": 0
}

furn_enc = {
    "unfurnished": 0,
    "semi furnished": 1,
    "furnished": 2,
    "unknown": 0
}

# Create new column by summing encoded values
df["condition_furnishing"] = (
    df["property_condition"].str.lower().map(prop_enc).fillna(0).astype(int) +
    df["furnishing"].str.lower().map(furn_enc).fillna(0).astype(int)
)

# Check results
print(df[["property_condition", "furnishing", "condition_furnishing"]].head())

# Optional: count distribution
print(df["condition_furnishing"].value_counts())

```
